```python
tweets_nuevos = []
```

```python
import re
for tweet in tweets:
    texto = tweet['tweet']
    usuario = tweet['usuario']
    texto.replace(usuario, "")
    matches = re.findall(r'@(\w+)', texto)

    tweets_nuevos.append({
        "usuario":matches,
        "tweet":texto
        

    })
```

```python
tweets_nuevos
```

Crear nueva base de datos

```python
import pandas as pd
df = pd.DataFrame(tweets_nuevos)
df.to_csv('tweets.csv', index=False)
df
```

Apilarlos

```python
# Leer el archivo CSV original
tweets = pd.read_csv('tweets.csv')
# Apilar los nuevos tweets al DataFrame original
tweets = pd.concat([tweets, tweets_nuevos], ignore_index=True)
# Imprimir el DataFrame resultante
print(tweets)
```
